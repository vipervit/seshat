```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Save as pickle:

```python
df_pops = pd.read_pickle('./populations.pkl')
df_pops
```

```python
df_china = df_pops[df_pops['NGA']=='Middle Yellow River Valley'].sort_values('From_Year')
df_india = df_pops[df_pops['NGA']=='Deccan'].sort_values('From_Year')
df_japan = df_pops[df_pops['NGA']=='Kansai'].sort_values('From_Year')
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(data=df_china, x='From_Year', y='Pop_Min', label='China')
sns.lineplot(data=df_india, x='From_Year', y='Pop_Min', label='India')
sns.lineplot(data=df_japan, x='From_Year', y='Pop_Min', label='Japan')
```
